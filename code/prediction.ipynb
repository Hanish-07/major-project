```python
df1 = pd.get_dummies(df1, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
df2 = pd.get_dummies(df2, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
df3 = pd.get_dummies(df3, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
df1[columns_to_scale] = standardScaler.fit_transform(df1[columns_to_scale])
df2[columns_to_scale] = standardScaler.fit_transform(df2[columns_to_scale])
df3[columns_to_scale] = standardScaler.fit_transform(df3[columns_to_scale])
```

```python
y1 = df1['target']
X1 = df1.drop(['target'], axis = 1)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size = 0.33, random_state = 0)

y2 = df2['target']
X2 = df2.drop(['target'], axis = 1)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size = 0.33, random_state = 0)

y3 = df3['target']
X3 = df3.drop(['target'], axis = 1)
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size = 0.33, random_state = 0)
```

K neighbours

```python
knn_scores1 = []
for k in range(1,31):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    knn_classifier.fit(X_train1, y_train1)
    knn_scores1.append(knn_classifier.score(X_test1, y_test1))

knn_scores2 = []
for k in range(1,31):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    knn_classifier.fit(X_train2, y_train2)
    knn_scores2.append(knn_classifier.score(X_test2, y_test2))

knn_scores3 = []
for k in range(1,31):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    knn_classifier.fit(X_train3, y_train3)
    knn_scores3.append(knn_classifier.score(X_test3, y_test3))
```

```python
plt.plot([k for k in range(1, 31)], knn_scores1, color = 'red')
for i in range(1,31):
    plt.text(i, knn_scores1[i-1], (i, knn_scores1[i-1]))
plt.xticks([i for i in range(1, 31)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
plt.plot([str(round(float(k), 2)) for k in range(1, 31)], knn_scores2, color = 'red')
for i in range(1,31):
    plt.text(i, knn_scores2[i-1], (i, knn_scores2[i-1]))
plt.xticks([i for i in range(1, 31)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
plt.plot([k for k in range(1, 31)], knn_scores3, color = 'red')
for i in range(1,31):
    plt.text(i, knn_scores3[i-1], (i, knn_scores3[i-1]))
plt.xticks([i for i in range(1, 31)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
print("The score for K Neighbors Classifier is {}% with {} nieghbors (Dataset 1).".format(knn_scores1[7]*100, 8))
print("The score for K Neighbors Classifier is {}% with {} nieghbors (Dataset 2).".format(knn_scores2[3]*100, 4))
print("The score for K Neighbors Classifier is {}% with {} nieghbors (Dataset 3).".format(knn_scores3[2]*100, 3))
```

Support Vector Classifier

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

svc_scores1 = []
for i in range(len(kernels)):
    svc_classifier = SVC(kernel = kernels[i])
    svc_classifier.fit(X_train1, y_train1)
    svc_scores1.append(svc_classifier.score(X_test1, y_test1))

svc_scores2 = []
for i in range(len(kernels)):
    svc_classifier = SVC(kernel = kernels[i])
    svc_classifier.fit(X_train2, y_train2)
    svc_scores2.append(svc_classifier.score(X_test2, y_test2))

svc_scores3 = []
for i in range(len(kernels)):
    svc_classifier = SVC(kernel = kernels[i])
    svc_classifier.fit(X_train3, y_train3)
    svc_scores3.append(svc_classifier.score(X_test3, y_test3))
```

```python
colors = rainbow(np.linspace(0, 1, len(kernels)))
plt.bar(kernels, svc_scores1, color = colors)
for i in range(len(kernels)):
    plt.text(i, svc_scores1[i], svc_scores1[i])
plt.xlabel('Kernels')
plt.ylabel('Scores')
plt.title('Support Vector Classifier scores for different kernels (Dataset 1)')
```

```python
plt.bar(kernels, svc_scores2, color = colors)
for i in range(len(kernels)):
    plt.text(i, svc_scores2[i], svc_scores2[i])
plt.xlabel('Kernels')
plt.ylabel('Scores')
plt.title('Support Vector Classifier scores for different kernels (Dataset 2)')
```

```python
plt.bar(kernels, svc_scores3, color = colors)
for i in range(len(kernels)):
    plt.text(i, svc_scores3[i], svc_scores3[i])
plt.xlabel('Kernels')
plt.ylabel('Scores')
plt.title('Support Vector Classifier scores for different kernels (Dataset 3)')
```

```python
print("The score for Support Vector Classifier is {}% with {} kernel (Dataset 1).".format(svc_scores1[0]*100, 'linear'))
print("The score for Support Vector Classifier is {}% with {} kernel (Dataset 2).".format(svc_scores2[1]*100, 'poly'))
print("The score for Support Vector Classifier is {}% with {} kernel (Dataset 3).".format(svc_scores3[0]*100, 'linear'))
```

Decision Tree 

```python
dt_scores1 = []
for i in range(1, len(X1.columns) + 1):
    dt_classifier = DecisionTreeClassifier(max_features = i, random_state = 0)
    dt_classifier.fit(X_train1, y_train1)
    dt_scores1.append(dt_classifier.score(X_test1, y_test1))

dt_scores2 = []
for i in range(1, len(X2.columns) + 1):
    dt_classifier = DecisionTreeClassifier(max_features = i, random_state = 0)
    dt_classifier.fit(X_train2, y_train2)
    dt_scores2.append(dt_classifier.score(X_test2, y_test2))

dt_scores3 = []
for i in range(1, len(X3.columns) + 1):
    dt_classifier = DecisionTreeClassifier(max_features = i, random_state = 0)
    dt_classifier.fit(X_train3, y_train3)
    dt_scores3.append(dt_classifier.score(X_test3, y_test3))
```

```python
plt.plot([i for i in range(1, len(X1.columns) + 1)], dt_scores1, color = 'green')
for i in range(1, len(X1.columns) + 1):
    plt.text(i, dt_scores1[i-1], (i, dt_scores1[i-1]))
plt.xticks([i for i in range(1, len(X1.columns) + 1)])
plt.xlabel('Max features')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different number of maximum features (Dataset 1)')
```

```python
plt.plot([i for i in range(1, len(X2.columns) + 1)], dt_scores2, color = 'green')
for i in range(1, len(X2.columns) + 1):
    plt.text(i, dt_scores2[i-1], (i, dt_scores2[i-1]))
plt.xticks([i for i in range(1, len(X2.columns) + 1)])
plt.xlabel('Max features')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different number of maximum features (Dataset 2)')
```

```python
plt.plot([i for i in range(1, len(X3.columns) + 1)], dt_scores3, color = 'green')
for i in range(1, len(X3.columns) + 1):
    plt.text(i, dt_scores3[i-1], (i, dt_scores3[i-1]))
plt.xticks([i for i in range(1, len(X3.columns) + 1)])
plt.xlabel('Max features')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different number of maximum features (Dataset 3)')
```

Random Forest

```python
estimators = [10, 100, 200, 500, 1000]

rf_scores1 = []
for i in estimators:
    rf_classifier = RandomForestClassifier(n_estimators = i, random_state = 0)
    rf_classifier.fit(X_train1, y_train1)
    rf_scores1.append(rf_classifier.score(X_test1, y_test1))

rf_scores2 = []
for i in estimators:
    rf_classifier = RandomForestClassifier(n_estimators = i, random_state = 0)
    rf_classifier.fit(X_train2, y_train2)
    rf_scores2.append(rf_classifier.score(X_test2, y_test2))

rf_scores3 = []
for i in estimators:
    rf_classifier = RandomForestClassifier(n_estimators = i, random_state = 0)
    rf_classifier.fit(X_train3, y_train3)
    rf_scores3.append(rf_classifier.score(X_test3, y_test3))
```

```python
colors = rainbow(np.linspace(0, 1, len(estimators)))
plt.bar([i for i in range(len(estimators))], rf_scores1, color = colors, width = 0.8)
for i in range(len(estimators)):
    plt.text(i, rf_scores1[i], rf_scores1[i])
plt.xticks(ticks = [i for i in range(len(estimators))], labels = [str(estimator) for estimator in estimators])
plt.xlabel('Number of estimators')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different number of estimators (Dataset 1)')
```

```python
plt.bar([i for i in range(len(estimators))], rf_scores2, color = colors, width = 0.8)
for i in range(len(estimators)):
    plt.text(i, rf_scores2[i], rf_scores2[i])
plt.xticks(ticks = [i for i in range(len(estimators))], labels = [str(estimator) for estimator in estimators])
plt.xlabel('Number of estimators')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different number of estimators (Dataset 2)')
```

```python
plt.bar([i for i in range(len(estimators))], rf_scores3, color = colors, width = 0.8)
for i in range(len(estimators)):
    plt.text(i, rf_scores3[i], rf_scores3[i])
plt.xticks(ticks = [i for i in range(len(estimators))], labels = [str(estimator) for estimator in estimators])
plt.xlabel('Number of estimators')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different number of estimators (Dataset 3)')
```

```python
print("The score for Random Forest Classifier is {}% with {} estimators (Dataset 1).".format(rf_scores1[1]*100, [100, 500]))
print("The score for Random Forest Classifier is {}% with {} estimators (Dataset 2).".format(rf_scores2[1]*100, [100, 200, 500]))
print("The score for Random Forest Classifier is {}% with {} estimators (Dataset 3).".format(rf_scores3[0]*100, [10]))
```

Comparing various algoriothms

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

raw_data = {'datasets': ['Dataset1', 'Dataset2', 'Dataset3'],
        'knn': [knn_scores1[7]*100,knn_scores1[3]*100,knn_scores1[2]*100],
        'svc': [svc_scores1[0]*100,svc_scores1[1]*100,svc_scores1[0]*100],
        'dt': [dt_scores1[17]*100,dt_scores1[27]*100,dt_scores1[2]*100],
        'rf': [rf_scores1[1]*100,rf_scores2[1]*100,rf_scores3[0]*100]}
df = pd.DataFrame(raw_data, columns = ['datasets', 'knn', 'svc', 'dt', 'rf'])

# Setting the positions and width for the bars
pos = list(range(len(df['datasets']))) 
width = 0.20 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with knn data,
# in position pos,
plt.bar(pos, 
        #using df['knn'] data,
        df['knn'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in datasets
        label=df['datasets'][0]) 

# Create a bar with svc data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['svc'] data,
        df['svc'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in datasets
        label=df['datasets'][1]) 

# Create a bar with dt data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['dt'] data,
        df['dt'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in datasets
        label=df['datasets'][2])

# Create a bar with dt data,
# in position pos + some width buffer,
plt.bar([p + width*3 for p in pos], 
        #using df['rf'] data,
        df['rf'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#000000', 
        # with label the third value in datasets
        label=df['datasets'][2])

# Set the y axis label
ax.set_ylabel('Scores')

# Set the chart's title
ax.set_title('Comparision of various algorithms with various datasets')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['datasets'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, 100] )

# Adding the legend and showing the plot
plt.legend(['KNN', 'SVC', 'DT', 'RF'], loc='upper right')
plt.grid()
plt.show()
```
